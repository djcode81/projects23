# **Lipinski descriptors**
Christopher Lipinski came up with a set of rule of thumb for evaluating the drug likeness of compounds. Druglikeness is based on (ADME), Absorption, Distribution, Metabolism, Excretion. It is also known as phamracokinetic profile.
AKA Rule-of-Five

RULE:

Molecular weight <500 Dalton

Octanol-water partition coefficient (LogP) < 5

Hydrogen bond donor < 5

Hydrogen bond acceptors < 10

```python
import numpy as np
from rdkit import Chem
from rdkit.Chem import Descriptors, Lipinski
```

**Calculuate descriptors**

```python
def lipinski(smiles, verbose=False):

    moldata= []
    for elem in smiles:
        mol=Chem.MolFromSmiles(elem)
        moldata.append(mol)

    baseData= np.arange(1,1)
    i=0
    for mol in moldata:

        desc_MolWt = Descriptors.MolWt(mol)
        desc_MolLogP = Descriptors.MolLogP(mol)
        desc_NumHDonors = Lipinski.NumHDonors(mol)
        desc_NumHAcceptors = Lipinski.NumHAcceptors(mol)

        row = np.array([desc_MolWt,
                        desc_MolLogP,
                        desc_NumHDonors,
                        desc_NumHAcceptors])

        if(i==0):
            baseData=row
        else:
            baseData=np.vstack([baseData, row])
        i=i+1

    columnNames=["MW","LogP","NumHDonors","NumHAcceptors"]
    descriptors = pd.DataFrame(data=baseData,columns=columnNames)

    return descriptors
```

```python
df_lipinski = lipinski(df.canonical_smiles)
```

## **Convert IC50 to pIC50 scale**
to allow IC50 data to be more uniformly distrivuted, IC50 will be converted to negative logarithmic scale to make the distribution more even.

```python
# https://github.com/chaninlab/estrogen-receptor-alpha-qsar/blob/master/02_ER_alpha_RO5.ipynb

import numpy as np

def pIC50(input):
    pIC50 = []

    for i in input['standard_value_norm']:
        molar = i*(10**-9) # Converts nM to M
        pIC50.append(-np.log10(molar))

    input['pIC50'] = pIC50
    x = input.drop('standard_value_norm', 1)

    return x
```

Cap the value to 100 million so that it wont have negative value (difficult for interpretation)

Create custom norm value that caps at 100 million

```python
def norm_value(input):
    norm = []

    for i in input['standard_value']:
        if i > 100000000:
          i = 100000000
        norm.append(i)

    input['standard_value_norm'] = norm
    x = input.drop('standard_value', 1)

    return x
```

```python
df_norm = norm_value(df_combined)
df_norm
```

```python
df_norm.standard_value_norm.describe()
```

```python
df_final = pIC50(df_norm)
df_final
```
